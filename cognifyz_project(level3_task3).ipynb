```python
import pandas as pd
import numpy as np
df = pd.read_csv('/content/Dataset .csv')
```

```python
price_delivery_table = df.groupby('Price range')[['Has Online delivery', 'Has Table booking']].value_counts(normalize=True) * 100
print("\nPercentage of Restaurants Offering Online Delivery & Table Booking:\n", price_delivery_table)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Bar plot for online delivery vs. price range
plt.figure(figsize=(8,5))
sns.barplot(x='Price range', y=df['Has Online delivery'].apply(lambda x: 1 if x == 'Yes' else 0), data=df)
plt.xlabel('Price Range')
plt.ylabel('Percentage Offering Online Delivery')
plt.title('Online Delivery by Price Range')
plt.show()

# Bar plot for table booking vs. price range
plt.figure(figsize=(8,5))
sns.barplot(x='Price range', y=df['Has Table booking'].apply(lambda x: 1 if x == 'Yes' else 0), data=df)
plt.xlabel('Price Range')
plt.ylabel('Percentage Offering Table Booking')
plt.title('Table Booking by Price Range')
plt.show()
```
